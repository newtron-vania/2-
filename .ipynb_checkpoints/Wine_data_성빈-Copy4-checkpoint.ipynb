```python
my_vectorizer = TfidfVectorizer(max_features = 12, stop_words = my_stopwords)
X = my_vectorizer.fit_transform(df3['cleaned']).toarray()   
print(my_vectorizer.get_feature_names())
```

```python
# 군집의 수를 키워가며 Sum of Squares Within을 계산한다.
ssw = []
cluster_ns = range(2,16)
for n in cluster_ns:
    print(n)
    my_cluster = KMeans(n,random_state=123)
    my_cluster.fit(X)
    ssw += [my_cluster.inertia_]         # SSW를 모아둔다.

```

```python
plt.figure(figsize = (12,6))
plt.plot(cluster_ns, ssw)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squares Within')
plt.show()
```

```python
# 군집의 수 = 4로 정하고 군집화를 실행한다.  <= 군집의 수 최적화가 선행되어야 한다! (부록 참고)
my_km = KMeans(n_clusters = 3, random_state = 123)
my_km.fit(X)
my_centroids = my_km.cluster_centers_               # 개개 군집의 중심점.
my_cluster_labels = my_km.labels_            
```

```python
pred=my_km.fit_predict(X)
```

```python
pd.DataFrame(X).groupby(pred).mean()
```

```python
for i in range(len(my_centroids)):
    print(i)
    print([my_vectorizer.get_feature_names()[x] for x in np.argsort(my_centroids[i])[-5:]])
```

```python
# PCA 차원축소 (2차원).
my_pca = PCA(n_components = 2)
transformed_comps = my_pca.fit_transform(X)         # Transformed 된 좌표.
df_transformed_comps = pd.DataFrame(data = transformed_comps, columns = ['PC1', 'PC2'])
df_transformed_comps=df_transformed_comps.join(pd.Series(my_cluster_labels, name='cluster_label'))
```

```python
# 산점도 시각화.
# my_colors = {0:'red',1:'blue',2:'green',3:'yellow'}    # 군집 label => Color.
my_names = {0: 'Spice', 1: 'Dry' ,
            2: 'Rich', 3: 'Moderate unknwons.'} # 군집 label => 명칭.

plt.figure(figsize = (6,6))
for a_cluster_n, df_small in df_transformed_comps.groupby('cluster_label'):
    plt.scatter('PC1', 'PC2', data = df_small, label = my_names[a_cluster_n], s = 10, alpha=0.6 )
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Clusters')
plt.legend(loc=4)
plt.show()
```

```python
from surprise import Reader                                                 # 리더 객체 생성해주는 클래스. 
from surprise import Dataset                                                # 알고리즘에 적합한 형태로 데이터를 가져다 주는 클래스.
from surprise.accuracy import rmse                                          # 평점 예측의 오류 (성능)을 계산해 주는 함수.
from surprise import SVD
# from surprise.prediction_algorithms.matrix_factorization import SVD         # SVD로 행렬분해 적용을 위한 클래스.
from surprise.model_selection import train_test_split                       # 학습용, 시험용 데이터 쪼개준는 함수.


my_reader = Reader(rating_scale=(80, 100))                     # 리더 객체 생성. 평점 레인지 설정.
my_data = Dataset.load_from_df(df=df3[['taster_name', 'title', 'points']], reader=my_reader)
type(my_data)
```

```python
data_train, data_test = train_test_split(my_data, test_size=0.3, shuffle=True)
```

```python
my_mf_algo = SVD(n_factors =50, n_epochs=20, verbose=True)             # n_factors = "은닉 요인(차원)". n_epochs = 경사하강법의 epoch 수.
my_mf_algo.fit(trainset=data_train)
```

```python
my_pred = my_mf_algo.test(testset=data_test)
rmse(predictions=my_pred)
```

```python
i_userId = 'Roger Voss'
predicted_ratings = []
for wine_name in df3['title'].unique():
    a_pred = my_mf_algo.predict(i_userId, wine_name)
    predicted_ratings.append((a_pred.est, wine_name))
```

```python
df_predicted_ratings = pd.DataFrame(predicted_ratings, columns = ['PredictedRating', "Movie"])
df_predicted_ratings.sort_values(by="PredictedRating", ascending = False, inplace=True)
df_predicted_ratings.reset_index(drop=True, inplace=True)
```
