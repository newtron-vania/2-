```python
import numpy as np
import re

import requests as rq
import bs4
import selenium
from selenium import webdriver

import konlpy

from sklearn.feature_extraction.text import CountVectorizer
import nltk

from selenium import webdriver
```

```python
#전처리
def cleaning(argStr):
    argStr = argStr.lower()
    argStr = re.sub(r'\W',  ' ',argStr)
    argStr = re.sub(r'\s+', ' ',argStr)
    argStr = re.sub(r'\s$', '' ,argStr)
    return(argStr)

#
def oneUrlNews(argLstHl):
    lstHyperLinkNews = []
    for hl in argLstHl:
        hlSubj = hl.find('strong')
        if hlSubj:
            strSubj = hlSubj.text
            #제목인지 판단
            if (len(strSubj) > 10):
                lstHyperLinkNews.append(hl)
                
    lstLinkNews = [x.get("href") for x in lstHyperLinkNews]
    
    link = lstLinkNews[np.random.randint(len(lstLinkNews))]
    if "https://news.naver.com" not in link:
        link = "https://news.naver.com" + link
        
    return(link)
```

```python
naverNewsUrl = "https://news.naver.com"

def oneTextNews(arUrl):
    driverMy = webdriver.Chrome('chromedriver.exe')
    
    driverMy.get("https://news.naver.com")
    html01 = driverMy.page_source
    soup01 = bs4.BeautifulSoup(html01, 'html.parser')
    lstHyperLink = soup01.find_all('a', class_=re.compile('clicks'))    
    urlTarget = oneUrlNews(lstHyperLink)
    
    driverMy.get(urlTarget)
    html02 = driverMy.page_source
    soup01 = bs4.BeautifulSoup(html02, 'html.parser')
    tagCont = soup01.find("div", id="articleBodyContents")
    
    return cleaning(tagCont.text)
    
strOneNews = oneTextNews(naverNewsUrl)
```

```python
strOneNews
```

```python
kkma = konlpy.tag.Kkma()
lstSent = kkma.sentences(strOneNews)
lstSent
```

```python
lstPosSent = []
for x in lstSent:
    lstPosSent.append(kkma.pos(x))
lstPosSent
```

```python
def makeLstSent(argLstPosSent, argOne=True):
    
    lstStop = ["E", "J"]
    lstStrSent = []
    for oneSent in argLstPosSent:
        strOneSent = ""
        for tp in oneSent:
            lstIsStop = [x in tp[1] for x in lstStop]
            if not any(lstIsStop):
                strOneSent += (" "+tp[0])
        lstStrSent.append(strOneSent)
    if argOne:
        lstStrSent = [" ".join(lstStrSent)]
    return(lstStrSent)
lstSent = makeLstSent(lstPosSent, True)
lstSent
```

```python
def makeNGram(argLstSent, argNum = 3):
    vectMy = CountVectorizer(ngram_range=(argNum, argNum), analyzer="word")
    lstNGram = vectMy.fit(argLstSent).get_feature_names()
    # ar2NGramCnt = vectMy.transform(argLstSent).toarray()        # 결과는 array of array.
    # lstNGramCnt = list(ar2NGramCnt[0])                          # 단순한 리스트로 변환.
    return(lstNGram)

lstNGram = makeNGram(lstSent)
lstNGram
```

```python
def makeBaseDic(argLstNGram, argNum=3):
    dicBase = {}
    for x in argLstNGram:
        lstWord = nltk.word_tokenize(x)
        preGram = ' '.join(lstWord[0:argNum-1])
        pstGram = lstWord[-1]
        
        if preGram not in dicBase.keys():
            dicBase[preGram] = [pstGram]
        else:
            dicBase[preGram] += [pstGram]
    return(dicBase)
dicBase = makeBaseDic(lstNGram)
dicBase
```

```python
def predict(argPreGram):
    rangeIdx = len(dicBase[argPreGram])
    idxPic = np.random.randint(0, rangeIdx)
    return(dicBase[argPreGram][idxPic])

predict("가방 목록")
```
